```python
df.columns = df.columns.str.strip()

df = df.rename(columns={
    "COBERTUTA 4G": "COBERTURA 4G"
})

df.columns
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
fig = px.histogram(df, x="target", title="Distribución del nivel de cobertura")
fig.show()
```

```python
tech_cols = [
    "COBERTURA 2G",
    "COBERTURA 3G",
    "COBERTURA HSPA+, HSPA+DC",
    "COBERTURA 4G",
    "COBERTURA LTE",
    "COBERTURA 5G"
]
```

```python
df["num_tecnologias"] = df[tech_cols].notnull().sum(axis=1)

df[["num_tecnologias"]].head()
```

```python
def clasificar_cobertura(x):
    if x >= 4:
        return "Alta"
    elif x >= 2:
        return "Media"
    else:
        return "Baja"

df["target"] = df["num_tecnologias"].apply(clasificar_cobertura)

df[["num_tecnologias", "target"]].head()
```

```python
fig = px.histogram(df, x="target", title="Distribución del nivel de cobertura")
fig.show()
```

```python
fig = px.histogram(df, x="PROVEEDOR", color="target", barmode="group")
fig.show()
```

```python
fig = px.box(df, x="DEPARTAMENTO", y="num_tecnologias", title="Cobertura por departamento")
fig.show()
```

```python
corr = df.select_dtypes(include="number").corr()

fig = px.imshow(corr, text_auto=True, title="Matriz de correlación")
fig.show()
```

```python
df_model = df[[
    "PROVEEDOR",
    "DEPARTAMENTO",
    "CABECERA MUNICIPAL",
    "num_tecnologias",
    "target"
]]

df_model.head()
```
